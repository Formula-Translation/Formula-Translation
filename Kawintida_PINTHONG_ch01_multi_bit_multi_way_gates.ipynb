## คัดลอกโค้ดของเกทจากแบบฝึกหัดที่แล้ว​

ให้คัดลอกนิยามอุปกรณ์สำหรับเกทในแบบฝึกหัดที่แล้ว (And, Or, Not, Xor, Mux และ DMux) มาวางไว้ในกล่องโค้ดด้านล่าง เพื่อให้สามารถนำเกทเหล่านี้มาใช้งานในแบบฝึกหัดนี้ได้

```
# Put your component codes from previous notebooks here

class Not(NotLayoutMixin, Component):
    IN = [w.In]
    OUT = [w.out]

    PARTS = [
      Nand(a=w.In, b=w.In, out=w.out),
    ]


class And(AndLayoutMixin, Component):
    IN = [w.a, w.b]
    OUT = [w.out]

    PARTS = [
        Nand(a=w.a, b=w.b, out=w.c),
        Not(In=w.c, out=w.out)
    ]


class Or(OrLayoutMixin, Component):
    IN = [w.a, w.b]
    OUT = [w.out]

    PARTS = [
        Nand(a=w.a, b=w.a, out=w.outA),
        Nand(a=w.b, b=w.b, out=w.outB),
        Nand(a=w.outB, b=w.outA, out=w.out),
    ]

class Xor(XorLayoutMixin, Component):
    IN = [w.a, w.b]
    OUT = [w.out]

    PARTS = [
        Nand(a=w.a, b=w.b, out=w.outAB),
        Nand(a=w.a, b=w.outAB, out=w.outABa),
        Nand(a=w.b, b=w.outAB, out=w.outABb),
        Nand(a=w.outABa, b=w.outABb, out=w.out),


    ]

class Mux(Component):
    IN = [w.a, w.b, w.sel]
    OUT = [w.out]

    PARTS = [
      Not(In=w.sel, out=w.notSel),
      Nand(a=w.notSel, b=w.a, out=w.outSela),
      Nand(a=w.sel, b=w.b, out=w.outSelb),
      Nand(a=w.outSela, b=w.outSelb, out=w.out),
    ]


class DMux(Component):
    IN = [w.In, w.sel]
    OUT = [w.a, w.b]

    PARTS = [
        Nand(a=w.sel, b=w.sel, out=w.outSel),
        Nand(a=w.outSel, b=w.In, out=w.outSelIn),
        Nand(a=w.outSelIn, b=w.outSelIn, out=w.a),
        Nand(a=w.sel, b=w.In, out=w.outSelIn2),
        Nand(a=w.outSelIn2, b=w.outSelIn2, out=w.b),

    ]
```

## Multi-bit gates: `Not16`, `And16`, `Or16`, `Mux16`

ตัวอย่างการทำงานแบบโต้ตอบ

* [Or16 gate](https://ecourse.cpe.ku.ac.th/courses/comsys/demo/or16.html)
* [Mux16 gate](https://ecourse.cpe.ku.ac.th/courses/comsys/demo/mux16.html)

ส่วนโค้ดสำหรับทดสอบจะอยู่ถัดจากการนิยามคลาสอุปกรณ์

```
# DO NOT ERASE THIS CELL - to be graded

class Not16(Component):
    IN = [w(16).In]
    OUT = [w(16).out]

    PARTS = [ # แนะนำให้สร้างเกท 16 ตัวด้วย for-loop

    ]
    for i in range(16):
      PARTS.append(Not(In=w.In[i], out=w.out[i]))
```

```
# DO NOT ERASE THIS CELL - to be graded

class And16(Component):
    IN = [w(16).a, w(16).b]
    OUT = [w(16).out]

    PARTS = [ # แนะนำให้สร้างเกท 16 ตัวด้วย for-loop
    ]
    for i in range(16):
      PARTS.append(And(a=w.a[i], b=w.b[i], out=w.out[i]))
```

```
# DO NOT ERASE THIS CELL - to be graded

class Or16(Component):
    IN = [w(16).a, w(16).b]
    OUT = [w(16).out]

    PARTS = [ # แนะนำให้สร้างเกท 16 ตัวด้วย for-loop
    ]
    for i in range(16):
      PARTS.append(Or(a=w.a[i], b=w.b[i], out=w.out[i]))
```

```
# DO NOT ERASE THIS CELL - to be graded

class Mux16(Component):
    IN = [w(16).a, w(16).b, w.sel]
    OUT = [w(16).out]

    PARTS = [ # แนะนำให้สร้างเกท 16 ตัวด้วย for-loop
    ]
    for i in range(16):
      PARTS.append(Mux(a=w.a[i], b=w.b[i], sel=w.sel, out=w.out[i]))
```

นิสิตสามารถทดลองและเทสเกทที่สร้างขึ้นโดยพิมพ์โค้ดสำหรับ trace, plot, interact ได้ที่ส่วนด้านล่างนี้

```
# พื้นที่ทดสอบ (ปรับเปลี่ยนตามต้องการ)

interact(Not16, depth=1)
interact(And16, depth=1)
interact(Or16, depth=1)
interact(Mux16, depth=1)
```

## Multi-way gates: `Or8Way`, `Mux4Way16`, `Mux8Way16`, `DMux4Way`, `DMux8Way`

ตัวอย่างการทำงานแบบโต้ตอบ

* [DMux4Way](https://ecourse.cpe.ku.ac.th/courses/comsys/demo/dmux4way.html)

ส่วนโค้ดสำหรับทดสอบจะอยู่ถัดจากการนิยามคลาสอุปกรณ์

```
# DO NOT ERASE THIS CELL - to be graded

class Or8Way(Component):
    IN = [w(8).In]
    OUT = [w.out]

    PARTS = [
        Or(a=w.In[0], b= w.In[1],out=w.out1),
        Or(a=w.In[2], b= w.In[3],out=w.out2),
        Or(a=w.In[4], b= w.In[5],out=w.out3),
        Or(a=w.In[6], b= w.In[7],out=w.out4),
        Or(a=w.out1, b= w.out2, out=w.out5),
        Or(a=w.out3, b= w.out4,out=w.out6),
        Or(a=w.out5, b= w.out6,out=w.out),



    ]
```

```
# DO NOT ERASE THIS CELL - to be graded

class Mux4Way16(Component):
    IN = [w(16).a, w(16).b, w(16).c, w(16).d, w(2).sel]
    OUT = [w(16).out]

    PARTS = [
        Mux16(a=w.a, b=w.b, sel=w.sel[0], out=w(16).out1),
        Mux16(a=w.c, b=w.d, sel=w.sel[0], out=w(16).out2),
        Mux16(a=w.out1, b=w.out2, sel=w.sel[1], out=w.out),

    ]
```

```
# DO NOT ERASE THIS CELL - to be graded

class Mux8Way16(Component):
    IN = [
        w(16).a,
        w(16).b,
        w(16).c,
        w(16).d,
        w(16).e,
        w(16).f,
        w(16).g,
        w(16).h,
        w(3).sel,
    ]
    OUT = [w(16).out]

    PARTS = [
        Mux4Way16(a=w.a, b=w.b, c=w.c, d=w.d, sel=w.sel[0:2],out=w(16).out1),
        Mux4Way16(a=w.e, b=w.f, c=w.g, d=w.h, sel=w.sel[0:2],out=w(16).out2),
        Mux16(a=w.out1, b=w.out2, sel=w.sel[2], out=w(16).out)
    ]
```

```
# DO NOT ERASE THIS CELL - to be graded

class DMux4Way(Component):
    IN = [w.In, w(2).sel]
    OUT = [w.a, w.b, w.c, w.d]

    PARTS = [
        DMux(In=w.In, sel= w.sel[1], a=w.out1 ,b= w.out2),
        DMux(In=w.out1, sel= w.sel[0], a=w.a ,b= w.b),
        DMux(In=w.out2, sel= w.sel[0], a=w.c ,b= w.d),


    ]
```

```
# DO NOT ERASE THIS CELL - to be graded

class DMux8Way(Component):
    IN = [w.In, w(3).sel]
    OUT = [w.a, w.b, w.c, w.d, w.e, w.f, w.g, w.h]

    PARTS = [
        DMux(In=w.In, sel= w.sel[2], a=w.out1 ,b= w.out2),
        DMux4Way(In=w.out1, sel= w.sel[0:2], a=w.a ,b= w.b, c= w.c, d=w.d),
        DMux4Way(In=w.out2, sel= w.sel[0:2], a=w.e ,b= w.f, c= w.g, d=w.h),

    ]
```

ใช้พื้นที่ด้านล่างทดสอบความถูกต้องของการสร้างอุปกรณ์
